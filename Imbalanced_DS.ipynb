This is Highly imbalanced DS


0--->Legal
1--->Illegal

```python
# seaparatin the legit and the farud transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

Undersampling method to handle imbalanced data

Build a sample DS containing similar distribution of legit and fraud transactions

Number of fraud Transactions =  241

```python
legit_sample = legit.sample(n=241)
```

```python
print(legit_sample.shape)
```

Concatenate the two dataframe


```python
new_dataset = pd.concat([legit_sample, fraud ], axis = 0)
```
