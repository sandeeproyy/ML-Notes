Matplotlib:
   Useful for Making Plots
  

```python
# importing the matplotlib library
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the Data

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
 # cosine wave
 plt.plot(x,z)
 plt.show
```

```python
# Labelling the Graph
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine Value')
plt.title('Sine Wave')
plt.show()
```

```python
# Plotting a Parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.ylabel('My Age')
plt.xlabel('Ages I would Fuck')
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show
```

```python
plt.plot(x,y, 'b^')
```

```python
plt.plot(x,y, 'rx')
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x,np.cos(x), 'r--')
plt.show()
```

Bar PLot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  #first two coordinates means the origin
languages = ['English','french','Spanish', 'Latin', 'German']
people = [100,50,150,40,60]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','french','Spanish', 'Latin', 'German']
people = [100,50,150,40,60]
ax.pie(people, labels=languages, autopct = '%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')     #ax.scatter is the main command here
ax.scatter(x,z, color='b')
plt.show()
```

3D Scatter Plot|

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c=z, cmap='Blues')
plt.show()
```
