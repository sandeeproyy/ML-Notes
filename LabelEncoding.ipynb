Labels to Numeric form

```python
# importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding DS 1

```python
# loading the DS from csv to pandas DF
cancer_data = pd.read_csv('/content/data.csv')
```

```python
cancer_data.head()
```

```python
cancer_data.shape
```

```python
# finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
# loading the LabelEncoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the DF
cancer_data['target'] =  labels
```

```python
cancer_data.head()
```

0 = benign
1 = malignant

```python
cancer_data['target'].value_counts()
```
